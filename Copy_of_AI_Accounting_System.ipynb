```python
# @title AI prompt cell

import ipywidgets as widgets
from IPython.display import display, HTML, Markdown,clear_output
from google.colab import ai

dropdown = widgets.Dropdown(
    options=[],
    layout={'width': 'auto'}
)

def update_model_list(new_options):
    dropdown.options = new_options
update_model_list(ai.list_models())

text_input = widgets.Textarea(
    placeholder='Ask me anything....',
    layout={'width': 'auto', 'height': '100px'},
)

button = widgets.Button(
    description='Submit Text',
    disabled=False,
    tooltip='Click to submit the text',
    icon='check'
)

output_area = widgets.Output(
     layout={'width': 'auto', 'max_height': '300px','overflow_y': 'scroll'}
)

def on_button_clicked(b):
    with output_area:
        output_area.clear_output(wait=False)
        accumulated_content = ""
        for new_chunk in ai.generate_text(prompt=text_input.value, model_name=dropdown.value, stream=True):
            if new_chunk is None:
                continue
            accumulated_content += new_chunk
            clear_output(wait=True)
            display(Markdown(accumulated_content))

button.on_click(on_button_clicked)
vbox = widgets.GridBox([dropdown, text_input, button, output_area])

display(HTML("""
<style>
.widget-dropdown select {
    font-size: 18px;
    font-family: "Arial", sans-serif;
}
.widget-textarea textarea {
    font-size: 18px;
    font-family: "Arial", sans-serif;
}
</style>
"""))
display(vbox)
```

### 1. System Initialization
This cell sets up the local directory structure for raw invoices, processed data, and initializes the SQLite database schema as defined in the implementation plan.

```python
import sqlite3
import os

# Create directories
os.makedirs('invoices/raw', exist_ok=True)
os.makedirs('invoices/processed', exist_ok=True)
os.makedirs('reports', exist_ok=True)

# Initialize Database
def init_db(db_path='accounting_system.db'):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Transactions table
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS transactions (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            date TEXT NOT NULL,
            description TEXT,
            amount REAL NOT NULL,
            category TEXT,
            source_file TEXT,
            status TEXT DEFAULT 'pending'
        )
    ''')

    # Audit logs for fraud/risk detection
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS audit_logs (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            transaction_id INTEGER,
            risk_score REAL,
            flags TEXT,
            timestamp DATETIME DEFAULT CURRENT_TIMESTAMP,
            FOREIGN KEY (transaction_id) REFERENCES transactions(id)
        )
    ''')

    conn.commit()
    conn.close()
    print(f"Database initialized at {db_path}")

init_db()
```

*italicized text*### 2. Data Ingestion Layer
This function handles adding new transactions to the database. It can be extended to process files or API inputs.

```python
def add_transaction(date, description, amount, category, source_file):
    conn = sqlite3.connect('accounting_system.db')
    cursor = conn.cursor()
    try:
        cursor.execute('''
            INSERT INTO transactions (date, description, amount, category, source_file)
            VALUES (?, ?, ?, ?, ?)
        ''', (date, description, amount, category, source_file))
        conn.commit()
        return cursor.lastrowid
    except Exception as e:
        print(f"Error: {e}")
        return None
    finally:
        conn.close()

# Example usage: Add a test record
transaction_id = add_transaction('2023-10-27', 'Cloud Services Subscription', 150.00, 'Technology', 'inv_001.pdf')
if transaction_id:
    print(f"Successfully added transaction with ID: {transaction_id}")
```

### 5. Unified Compliance View
This report merges transaction details with audit results to help identify high-risk items quickly.

```python
def get_compliance_report():
    conn = sqlite3.connect('accounting_system.db')
    query = '''
        SELECT t.id, t.date, t.description, t.amount, t.category,
               a.risk_score, a.flags
        FROM transactions t
        LEFT JOIN audit_logs a ON t.id = a.transaction_id
    '''
    df = pd.read_sql_query(query, conn)
    conn.close()
    return df

# Display compliance report
compliance_df = get_compliance_report()
compliance_df
```

### 7.1 LLM-Style Structured Data Extraction
Instead of hardcoded values, this module uses pattern matching (simulating an LLM's extraction capabilities) to pull actual data from the raw OCR strings.

```python
import re
from datetime import datetime

def extract_structured_data(text):
    """
    Simulates an LLM parsing raw text to find accounting entities.
    """
    # 1. Try to find an amount (e.g., $123.45 or 123.45)
    amount_match = re.search(r'\$?\s?(\d+[.,]\d{2})', text)
    amount = float(amount_match.group(1).replace(',', '')) if amount_match else 0.0

    # 2. Try to find a date (e.g., YYYY-MM-DD or MM/DD/YYYY)
    date_match = re.search(r'(\d{4}-\d{2}-\d{2}|\d{2}/\d{2}/\d{4})', text)
    date_str = date_match.group(1) if date_match else datetime.now().strftime('%Y-%m-%d')

    # 3. Clean description (first line or first few words)
    description = text.split('\n')[0][:50].strip()
    if not description:
        description = "Unknown Document"

    return {
        'date': date_str,
        'description': f"Parsed: {description}",
        'amount': amount,
        'category': 'Parsed-Pending'
    }

# Let's test it on a string that looks more like a real invoice
sample_invoice_text = """
INVOICE #12345
Date: 2023-12-15
Vendor: Acme Corp
Total Amount: $1,250.00
Services: Consulting
"""

structured_info = extract_structured_data(sample_invoice_text)
print("Extracted Data:", structured_info)

# Integrate back to database
new_id = add_transaction(
    structured_info['date'],
    structured_info['description'],
    structured_info['amount'],
    structured_info['category'],
    'simulated_invoice.txt'
)

if new_id:
    run_audit(new_id, structured_info['amount'])
    print(f"Created Transaction {new_id} from parsed text.")
```

### 8. Audit Analytics: Benford's Law Analysis
In accounting data science, Benford's Law predicts the frequency of leading digits. Significant deviations can signal manual intervention or fraud.

```python
import matplotlib.pyplot as plt
import numpy as np

def benford_analysis(df, column):
    # Extract the first digit of each value
    digits = df[column].abs().astype(str).str[0].astype(int)
    digits = digits[digits > 0]

    # Calculate observed frequencies
    observed_freq = digits.value_counts(normalize=True).sort_index()

    # Benford's Law expected frequencies
    expected_freq = pd.Series({d: np.log10(1 + 1/d) for d in range(1, 10)})

    # Plotting
    plt.figure(figsize=(10, 5))
    plt.bar(observed_freq.index - 0.2, observed_freq, width=0.4, label='Observed', color='skyblue')
    plt.bar(expected_freq.index + 0.2, expected_freq, width=0.4, label='Expected (Benford)', color='orange', alpha=0.7)
    plt.xticks(range(1, 10))
    plt.xlabel('Leading Digit')
    plt.ylabel('Frequency')
    plt.title(f'Benford\'s Law Analysis on {column}')
    plt.legend()
    plt.show()

# Adding more dummy data to make the analysis meaningful
for i in range(10):
    add_transaction('2023-11-02', f'Test Expense {i}', np.random.uniform(10, 1000), 'General', 'batch_test.csv')

all_data = get_all_transactions()
benford_analysis(all_data, 'amount')
```
